```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'D:\downloads\orders.csv'
orders_df = pd.read_csv('D:\downloads\orders.csv')

# Rename columns for better readability
orders_df.rename(columns={'cust id': 'cust_id', 'prod id': 'prod_id'}, inplace=True)

# Convert the 'date' column to datetime format
orders_df['date'] = pd.to_datetime(orders_df['date'], format='%d.%m.%Y')

# Plot the scatter plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(orders_df['date'], orders_df['quantity'], c=orders_df['cust_id'], cmap='viridis')
plt.colorbar(label='Customer ID')
plt.title('Order Quantity Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.grid(True)
plt.show()

# Plot the scatter plot using Seaborn for a more aesthetic visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(data=orders_df, x='date', y='quantity', hue='cust_id', palette='viridis', s=100)
plt.title('Order Quantity Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend(title='Customer ID')
plt.grid(True)
plt.show()
```

```python

# Rename columns for better readability
orders_df.rename(columns={'cust id': 'cust_id', 'prod id': 'prod_id'}, inplace=True)

# Convert the 'date' column to datetime format
orders_df['date'] = pd.to_datetime(orders_df['date'], format='%d.%m.%Y')

# Since 'date' is not a numerical column, we should drop it for correlation matrix computation
correlation_matrix = orders_df.drop(columns=['date']).corr()

# Print the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
