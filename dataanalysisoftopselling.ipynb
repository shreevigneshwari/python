```python
import pandas as pd

# Load the datasets
customers_df = pd.read_csv('D:\downloads\customers.csv')  # Assuming you have a CSV file named 'customers.csv'
products_df = pd.read_csv('D:\downloads\products.csv')  # Assuming you have a CSV file named 'products.csv'
  # Assuming you have a CSV file named 'ratings.csv'
orders_df = pd.read_csv('D:\downloads\orders.csv')  # Assuming you have a CSV file named 'orders.csv'

# Assuming 'orders_df' contains order details with 'product_id' and 'quantity' columns
# Merge orders with products to get product details
merged_df = pd.merge(orders_df, products_df, on='prod id', how='left')

# Aggregate sales quantity for each product
product_sales = merged_df.groupby('prod id')['quantity'].sum().reset_index()

# Merge product sales with product details
product_performance = pd.merge(product_sales, products_df, on='prod id', how='left')

# Sort products by sales quantity in descending order
top_performing_products = product_performance.sort_values(by='quantity', ascending=False)

# Display top performing products
print("Top Performing Products:")
print(top_performing_products.head(10))  # Display top 10 performing products
```

```python
import pandas as pd

# Load the datasets
customers_df = pd.read_csv('D:\downloads\customers.csv')  # Assuming you have a CSV file named 'customers.csv'
products_df = pd.read_csv('D:\downloads\products.csv')  # Assuming you have a CSV file named 'products.csv'
orders_df = pd.read_csv('D:\downloads\orders.csv')  # Assuming you have a CSV file named 'orders.csv'

# Assuming 'orders_df' contains order details with 'product_id' and 'quantity' columns
# Merge orders with products to get product details
merged_df = pd.merge(orders_df, products_df, on='prod id', how='left')

# Calculate revenue for each product
merged_df['revenue'] = merged_df['quantity'] * merged_df['price']

# Aggregate revenue for each product
product_revenue = merged_df.groupby('prod id')['revenue'].sum().reset_index()

# Merge product revenue with product details
product_performance = pd.merge(product_revenue, products_df, on='prod id', how='left')

# Sort products by revenue in descending order
top_revenue_products = product_performance.sort_values(by='revenue', ascending=False)

# Display top revenue products
print("Top Revenue Products:")
print(top_revenue_products.head(10))  # Display top 10 revenue products
```
