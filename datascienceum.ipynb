```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'greendestination.csv'
data = pd.read_csv(file_path)

# Check for missing values in Age and Attrition columns
missing_values = data[['Age', 'Attrition']].isnull().sum()
print("Missing values:\n", missing_values)

# Convert the Attrition column to numerical values
data['Attrition'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Statistical summary of age for employees who have left and those who have not
age_attrition_summary = data.groupby('Attrition')['Age'].describe()
print("\nStatistical summary of age by attrition status:\n", age_attrition_summary)

# Visualization: Age distribution for employees who have left and those who have not
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='Age', hue='Attrition', multiple='stack', bins=20)
plt.title('Distribution of Age by Attrition Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Attrition', labels=['No', 'Yes'])
plt.show()
```

```python

# Check for missing values in YearsAtCompany and Attrition columns
missing_values_years_at_company = data[['YearsAtCompany', 'Attrition']].isnull().sum()
print("Missing values:\n", missing_values_years_at_company)

# Convert the Attrition column to numerical values if not already converted
data['Attrition'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Statistical summary of YearsAtCompany for employees who have left and those who have not
years_at_company_attrition_summary = data.groupby('Attrition')['YearsAtCompany'].describe()
print("\nStatistical summary of years at company by attrition status:\n", years_at_company_attrition_summary)

# Visualization: Years at company distribution for employees who have left and those who have not
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='YearsAtCompany', hue='Attrition', multiple='stack', bins=20)
plt.title('Distribution of Years at Company by Attrition Status')
plt.xlabel('Years at Company')
plt.ylabel('Count')
plt.legend(title='Attrition', labels=['No', 'Yes'])
plt.show()
```

```python

# Check for missing values in MonthlyIncome and Attrition columns
missing_values_income = data[['MonthlyIncome', 'Attrition']].isnull().sum()
print("Missing values:\n", missing_values_income)

# Convert the Attrition column to numerical values if not already converted
data['Attrition'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Statistical summary of MonthlyIncome for employees who have left and those who have not
income_attrition_summary = data.groupby('Attrition')['MonthlyIncome'].describe()
print("\nStatistical summary of income by attrition status:\n", income_attrition_summary)

# Visualization: Monthly income distribution for employees who have left and those who have not
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='MonthlyIncome', hue='Attrition', multiple='stack', bins=20)
plt.title('Distribution of Monthly Income by Attrition Status')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.legend(title='Attrition', labels=['No', 'Yes'])
plt.show()
```
