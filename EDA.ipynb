```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
file_path = 'D:\downloads\country.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Pie chart for Languages
language_counts = data['Language'].value_counts()
plt.figure(figsize=(8, 8))
language_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(language_counts))))
plt.title('Distribution of Languages')
plt.ylabel('')
plt.show()

# Pie chart for Countries
country_counts = data['Country'].value_counts()
plt.figure(figsize=(8, 8))
country_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(country_counts))))
plt.title('Distribution of Countries')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
file_path = 'D:\downloads\country.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Create histograms for all columns
for column in data.columns:
    plt.figure(figsize=(10, 6))
    if data[column].dtype == 'object':
        # Categorical data
        data[column].value_counts().plot(kind='bar', color='skyblue')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {column}')
    else:
        # Numerical data
        data[column].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {column}')
    
    plt.show()
```
